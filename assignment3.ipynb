Consider the vector potential if the direction of propagating is along the z axis,

$$\vec{A}(\omega; \vec{r}, t)=2\vec{A_0}(\omega)\,cos(\vec{k}\cdot\vec{r}-\omega t+\delta_\omega)$$

Writing it as,

$$A_x^L= A_x^R=\sqrt{2}\vec{A_0}(\omega)\,cos(k\cdot z-\omega t+\delta_\omega)$$
$$A_y^L= -A_y^R=\sqrt{2}\vec{A_0}(\omega)\,sin(k\cdot z-\omega t+\delta_\omega)$$
$$A_z^L=A_z^R=0$$

The corresponding electric field vectors are

$$E_x^L= E_x^R=-\sqrt{2}\omega\vec{A_0}(\omega)\,sin(k\cdot z-\omega t+\delta_\omega)$$
$$E_y^L= -E_y^R=\sqrt{2}\omega\vec{A_0}(\omega)\,cos(k\cdot z-\omega t+\delta_\omega)$$
$$E_z^L=E_z^R=0$$

From the direction of $\vec{k}$, the vector $\vec{E}^L$ is seen to be of constant magnitude but rotating in an anticlockwise direction in the $xy$ plane with frequency $\omega$ (left circularly polarized light) while the vector $\vec{E}$^R is of constant magnitude but rotates in a clockwise direction with frequency $\omega$ (right circularly polarized light). 

$$\vec{A^L}=\vec{A_0}(\omega)\left[ \hat{e}^L e^{i(k\cdot z-\omega t+\delta_\omega)} + e^{-i(k\cdot z-\omega t+\delta_\omega)}\right]$$
$$\vec{A^R}=\vec{A_0}(\omega)\left[ \hat{e}^R e^{i(k\cdot z-\omega t+\delta_\omega)} + e^{-i(k\cdot z-\omega t+\delta_\omega)}\right]$$

where $\hat{e}^L$ and $\hat{e}^R$ are complex orthogonal unit vectors given by

$$\hat{e}^L = \frac{1}{\sqrt{2}} (\hat{x}+i\hat{y})$$
$$\hat{e}^R = \frac{1}{\sqrt{2}} (\hat{x}-i\hat{y})$$

The term in $\vec{A}$ associated with $e^{i(k\cdot z-\omega t)}$ corresponds to the absorption of photons while the term $e^{-i(k\cdot z-\omega t)}$ to the emission of photons. So the term $\hat{e}^L\cdot \vec{r_{ab}}$ corresponds to the absorption of left circularly polarized light while $\hat{e}^R\cdot \vec{r_{ab}}$ corresponds to the absorption of right circularly polarized light. Similarly, $\hat{e}^{L*}\cdot \vec{r_{ab}^*}=\hat{e}^{L*}\cdot \vec{r_{ab}}$ corresponds to the emission of left circularly polarized light and $\hat{e}^{R*}\cdot \vec{r_{ab}^*}=\hat{e}^{R*}\cdot \vec{r_{ab}}$ corresponds to the emission of right circularly polarized light.

Therefore,

$$\hat{e}^L\cdot \vec{r_{ab}} = \frac{1}{\sqrt{2}} (\hat{x}+i\hat{y})\cdot \vec{r_{ab}} = \frac{1}{\sqrt{2}} (\hat{x_{ab}}+i\hat{y_{ab}})$$

Now, the matrix elements of the matrices $\hat{x}$ and $\hat{y}$ are all zero. So, 

$$\hat{e}^L\cdot \vec{r_{ab}} = 0$$

Therfore, no transitions occur among $\psi_1(\vec{r})$, $\psi_2(\vec{r})$ or $\psi_3(\vec{r})$ under the incidence of left or right circularly polarized light.

```python
import numpy as np
from numpy import linalg as LA

n=5
A=np.zeros((n, n), float)
E=10
B=5
for i in range(0, n):
	A[i][i]=E
for i in range(0, n-1):
    A[i][i+1]=B
for i in range(0, n-1):
    A[i+1][i]=B
w, v = LA.eig(A)

print("The Hamiltonian matrix:")
for i in range(0, n):
	print(A[i])
print()
print("The energy eigenvalues:")
for i in range(0, n):
    print(w[i])
print()
print("Eigenvectors:")
for i in range(0, n):
	print(v[i])
x=np.array([[24, 0, 0, 0, 9/2],
		   [0, -24, -9/2, 0, 0],
		   [0, -9/2, -12, -3/2, 0],
		   [0, 0, -3/2, 0, 3/2],
		   [9/2, 0, 0, 3/2, 12]])
r=np.zeros((n, n))
for i in range(0, n):
	for j in range(0, n):
		r[i][j]=LA.multi_dot([v[i], x, v[j]])
print()
print("The transition dipole moment matrix:")
for i in range(0, n):
	print(r[i])
```

The expression for $c(\omega, t)$ is given by

$$ c_a(\omega, t)=-\frac{e}{m} A_0(\omega) M_{ab}(\omega) e^{-i\delta_\omega} \int_{0}^{t} e^{[i(\omega-\omega_{ab})t'-\frac{t'}{2 \tau_b}]} dt'$$
$$ = -\frac{e}{m} A_0(\omega) M_{ab}(\omega) e^{-i\delta_\omega} \left[ \frac{e^{[i(\omega-\omega_{ab})t-\frac{t}{2 \tau_b}]}-1}{i(\omega-\omega_{ab})-\frac{1}{2 \tau_b}} \right] $$

At times $t>>\tau_b$, the probablity that a photon has been emitted is given by $|c(\omega, t)|^2$ which is proportional to

$$ \left| \frac{M_{ab}(\omega)}{i(\omega-\omega_{ab})-\frac{1}{2 \tau_b}} \right|^2 = \frac{(M_{ab}(\omega))^2}{(\omega-\omega_{ab})^2+\frac{1}{4 \tau_b^2}}$$

The integral has a shrp maximum at $\omega-\omega_{ab}=0$ or $\omega=\omega_{ab}$. So, $\omega$ is set $\omega=\omega_{ab}$ in the slowly varying quantities $A_0^2(\omega)$ and $|M_{ab}(\omega)|^2$ giving $A_0^2(\omega_{ab})$ and $|M_{ab}(\omega_ab)|^2$.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"]=15,24

omega=np.arange(-100.0, 500.0, 0.1)
omega_21=56
m_21=-15.59
tau_2=10
omega_41=208
m_41=-0.84
tau_4=0.1
omega_51=263
m_51=-17.34
tau_5=0.01
intensity2=np.square(m_21)/(np.square(omega-omega_21)+np.square(0.5*tau_2))
intensity4=np.square(m_41)/(np.square(omega-omega_41)+np.square(0.5*tau_4))
intensity5=np.square(m_51)/(np.square(omega-omega_51)+np.square(0.5*tau_5))
fig, ax = plt.subplots(3)
ax[0].plot(omega, intensity2, color='blue')
ax[1].plot(omega, intensity4, color='blue')
ax[2].plot(omega, intensity5, color='blue')
ax[0].set(xlabel='$\omega$(in $10^{14}$ Hz)', ylabel='$|c(\omega, t)|^2$',
			 title=r'Intensity distribution for 2 to 1 transitions($\tau_2=10^{-15}$ seconds)')
ax[1].set(xlabel='$\omega$(in $10^{14}$ Hz)', ylabel='$|c(\omega, t)|^2$',
			 title=r'Intensity distribution for 4 to 1 transitions($\tau_4=10^{-13}$ seconds)')
ax[2].set(xlabel='$\omega$(in $10^{14}$ Hz)', ylabel='$|c(\omega, t)|^2$',
			 title=r'Intensity distribution for 5 to 1 transitions($\tau_5=10^{-12}$ seconds)')
plt.show()
```

The plots have been generated separately bacuuse of the wide variation in the order of magnitude of $|c(\omega, t)|^2$. As, $|c(\omega, t)|^2 \tilde \tau_b^2$, and $\tau_b^2$ varies over 3 order of magnitude, $|c(\omega, t)|^2$ varies over $2\times 3=6$ orders of magnitude as it can be inderred from the ordinate of the graphs. So, the graphs have been generated separately, so that the plot for the highest order of magnitude($10^{-12}$ for the transition from 5 to 1) does not over power the plot for the lower order of magnitude($10^{-15}$ for the transition from 2 to 1).

Using the uncertainty relation, $\Delta E\cdot \Delta t \geq \frac{\hbar}{2}$, the shorter is the lifetime $\tau$ of a state, the greater would be its line width

$$ \Delta\omega = \frac{\Delta E}{\hbar} = \frac{1}{\Delta t} = \frac{1}{\tau}$$

So, as the lifetime of the states increases in the order $10^{-16}$ s, $10^{-15}$ s to $10^{-13}$ s, the spectral lines become sharper, which is observed in the plots above. 

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"]=15,8

omega=np.arange(-100.0, 500.0, 0.1)
omega_32=76
m_32=-12.12
tau_3=1
intensity=np.square(m_32)/(np.square(omega-omega_32)+np.square(0.5*tau_3))
fig, ax = plt.subplots()
ax.plot(omega, intensity, color='blue')
ax.set(xlabel='$\omega$(in $10^{14}$ Hz)', ylabel='$|c(\omega, t)|^2$',
			 title=r'Intensity distribution for 3 to 2 transitions($\tau_3=10^{-14}$ seconds)')
plt.show()
```
