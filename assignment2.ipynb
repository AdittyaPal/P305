Consider energy of the electron in such a state to be $-10\,eV$. Consider a molecule formed by two such atoms at proximity $1.5\mathring A$ in the X direction. Let be $-5\,eV$ which is generally reasonable for 2s electrons. Consider excitation of an electron from the ground state of the two atom system to the available excited state (as you saw in assignment 1 A). Note that there you got only one excited state because you consider a basis made of only two states (approximated to be orthonormal). In principle there are of course infinite number of excited states. Calculate $|R_{21}|$, $\omega_{21}$ and $W^D_{21}$ estimate transition rate we derived in the class and make an estimate of number of transitions which may be occurring in a second at room temperature.
Note that in deriving expression for in class we have populated the entire phase space with
photons. So we are essentially talking of the maximum possible values.

$$|r_{21}|=<\psi_2|\hat r|\psi_1>$$
As $\psi_1$ and $\psi_2$ are constant functions the calculation of $|r_{21}|$ reduces to the integral
$$\int_{x=0.25}^{x=1.75}\int_{y=-1}^{y=1}\int_{z=-1}^{z=1} \sqrt{x^2+y^2+z^2} \,dx\,dy\,dz$$
The integral has been evaluated by the code

```python
import math
from scipy import integrate
x=0.25
dx=0.01
dy=0.01
dz=0.01
sum=0.0
while(x<1.75):
    y=-1.0
    while(y<1.0):
        z=-1.0
        while(z<1.0):
            sum+=math.sqrt(x**2+y**2+z**2)*dx*dy*dz
            z+=dz
        y+=dy
    x+=dx
print("Integral: "+str(sum))

f=lambda z, y, x: math.sqrt(x**2+y**2+z**2)
result=integrate.tplquad(f, 0.25, 1.75, lambda x: -1, lambda x: 1, lambda x, y: -1, lambda x, y: 1)
print(result)
```

For the given system, how much time of exposure to the appropriate electromagnetic radiation is required for the population of the excited and ground states to be equal if there are only the stimulated processes (emission as well as absorption) in action at room temperature. Consider of course that the system was entirely in the ground state before exposure to the required electromagnetic field.

For an ensemble of states, let $N_1$ of them be in the lower state $\psi_1$ with energy $E_1=-15\,eV$ and $N_2$ of them in the upper state $\psi _2$ with energy $E_2=-5\,eV$. Let $W_{21}$ be the absorption rate. The number of upward transitions taking place by absorption is proportional to the population of the lower state $\psi_1$, $N_1$ and is $W_{21}N_1$. Similarly, the number number of downward transitions by stimulated emission is proportional to the population of the upper state $\psi_2$, $N_2$ and is $W_{12}N_2$, where $W_{12}$ is the rate of stumilated transitions from the state $\psi_2$ to $\psi_1$. This sets up the two differential equations,
$$\frac{dN_1}{dt}=-W_{21}N_1+W_{12}N_2$$
$$\frac{dN_2}{dt}=W_{21}N_1-W_{12}N_2$$
Solving these differential equations with the initial conditions,
$$N_1=c_1=1;\;\;N_2=c_2=0$$
by the following code gives

```python
import matplotlib.pyplot as plt
import numpy as np

t=np.arange(0.0, 5000.0, 1.0)
c_1=np.zeros(5000)
c_2=np.zeros(5000)
c_1[0]=1
rate=7.0e-4
i=1
while(i<5000):
	c_1[i]=c_1[i-1]-c_1[i-1]*rate+c_2[i-1]*rate
	c_2[i]=c_2[i-1]+c_1[i-1]*rate-c_2[i-1]*rate
	i+=1
print(c_1[4999])
print(c_2[4999])
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(t, c_1, color='blue', label='$c_1$')
ax.plot(t, c_2, color='green', label='$c_2$')
ax.set(xlabel='Time(in $10^{-12}$ seconds)', ylabel='$c_i$', title='Time Evolution of the two atomic states')
ax.legend()
#fig.savefig("timeEvol.png")
plt.show()
```
